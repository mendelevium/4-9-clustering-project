<h1>Data Cleaning</h1>

```python
import pandas as pd
import json
import numpy as np
```

<h2>Data Place Clean</h2>

```python
def clean_opening_hours(x):
    if type(x) != float:
        x = (x.replace("'", '"')
             .replace(': ', ':"')
             .replace('}', '"}'))
        x = json.loads(x)
        x = x['open_now']
        if x == "True":
            x = True
        elif x == "False":
            x = False
    return x
```

```python
df_place = pd.read_csv('data/data_place_raw.csv')
df_place = df_place.drop('Unnamed: 0', axis = 1)

#Cleaning opening_hours to open_now
df_place['opening_hours'] = df_place['opening_hours'].apply(clean_opening_hours)
df_place = df_place.rename(columns = {'opening_hours': 'open_now'})

#Remove duplicate place_id rows
df_place = df_place.drop_duplicates(subset = ['place_id'])

#Save file to csv
df_place.to_csv('data/data_place_clean.csv')
```

<h2>Data Reviews Clean</h2>

```python
dfs_reviews = []
for i in range(10):
    file = f'data/data_reviews_{i}.csv'
    df = pd.read_csv(file)
    dfs_reviews.append(df)
    
#Merge batches of csv files
df_reviews = pd.concat(dfs_reviews, ignore_index = True)

#Drop Unnamed column
df_reviews = df_reviews.drop('Unnamed: 0', axis = 1)

#Drop duplicates columns with same text
df_reviews = df_reviews.drop_duplicates(subset = ['text'])

#Save file to csv
df_reviews.to_csv('data/data_reviews_clean.csv')
```
